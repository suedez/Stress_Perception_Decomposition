```python
new_df = vignette_dat.merge(objective_code, how='left', on='vid')
```

```python
new_df=new_df[new_df.score!='']
```

```python
new_df.score.unique()
```

```python
new_df.groupby(['Session','Content Classification (1-7)','score']).size()
```

```python
new_df2= pd.DataFrame({'count' : new_df.groupby(['Session','Content Classification (1-7)','score']).size()}).reset_index()
new_df2
```

```python
new_df2['Content Classification (1-7)'] = new_df2['Content Classification (1-7)'].replace([1,2,3,4,5,6,7],['Interpersonal Tension','Work/Education','Home','Finances','Health/Accident','Discrimination','Misc'])
```

```python
df = new_df2.pivot_table(values='count', index=['Session','Content Classification (1-7)'], columns='score', aggfunc='first')
df
```

```python
df.to_csv('./DataTransformation.csv', sep=',', header=True, index=True)
```
